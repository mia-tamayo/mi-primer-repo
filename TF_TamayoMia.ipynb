# Trabajo Final: Análisis de Percepciones sobre la Democracia en el Perú


El propósito del presente trabajo es responder a nivel descriptivo la siguiente pregunta de investigación: ¿Qué tan buena es la percepción ciudadana al respecto del funcionamiento de la democracia en el Perú?

Para ello, se utilizará la base "Actitudes hacia la Democracia" (2023) proveniente del Observatorio para la Democracia y Gobernabilidad de la Pontificia Universidad Católica del Perú.

```python
#Primero cargamos las librerías y paquetes a utilizar
import pandas as pd
import altair as alt
```

Ahora cargamos nuestra base de datos

```python
#Leemos nuestra base: Funcionamiento de la democracia a nivel regional (2023)
locacion_democracia = "/content/Funcionamiento de la democracia - Regional (2023) (1).xlsx"
df = pd.read_excel(locacion_democracia)
df
```

Para la sección de tratamiento de datos, vamos a crear una función para agrupar los 24 departamentos a nivel de macroregiones

```python
def asignar_macroregion(region):
  if region in ["Ancash","La Libertad","Piura","Cajamarca","Lambayeque","Tumbes"]:
    return "Norte"
  elif region in ["Arequipa","Apurímac","Puno","Moquegua","Cusco","Tacna"]:
    return "Sur"
  elif region in ["Ica","Junín","Ayacucho","Pasco","Huancavelica","Huánuco"]:
    return "Centro"
  elif region in ["San Martín","Amazonas","Loreto","Ucayali","Madre de Dios"]:
    return "Selva"
  else:
    return "Resto del Perú"

df['Macroregiones']=df['Región'].apply(asignar_macroregion)
df
```

Un paso previo al análisis descriptivo, procedemos a realizar un gráfico de barras a nivel de macroregiones para analizar la frecuencia de qué tan buena consideran la democracia los ciudadanos acorde a la ubicación geográfica en la que viven

```python
alt.Chart(df).mark_bar()
```

Ahora procedemos a explicar posiciones para nuestros puntos (los ejes x e y)

```python
alt.Chart(df).mark_bar().encode(
    y=alt.Y("Macroregiones"),
    x=alt.X('Bien - Muy bien:Q', title='Funcionamiento: Bien - Muy Bien'),
    color = "Macroregiones:N"
)
```

### Análisis descriptivo:
Para este breve proyecto se deseó realizar una comparativa a nivel macroregional sobre el funcionamiento de la democracia en el Perú. Luego de agrupar los departamentos a nivel de macro regiones y habiendo realizado el gráfico de barras respectivo, algunas consideraciones:

* La percepción ciudadana para el año 2023 en el resto del Perú (incluye Lima Metropolitana - la capital del país - Lima Provincias y Callao) ha obtenido el nivel más bajo en comparación con sus demás regiones, siendo un objeto de análisis interesante. ¿A qué se debe esta diferenciación en la capital versus los demás departamentos?
* En su contra-parte, la macro región Norte posee un nivel diferenciado un poco más alto sobre el funcionamiento de la democracia, encabezando en segundo lugar la macro región Selva.

* Tanto la macro región Centro como Sur presentan resultados descriptivos estadísticamente similares, no superando la cifra de 1.5 a diferencia de las otras categorías.
